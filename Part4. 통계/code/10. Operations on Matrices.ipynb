```python
# 행렬의 곱셈의 코드 표현 실습

D = torch.tensor([[1, 1, 3],
                  [4, 5, 6],
                  [7, 8, 9]])

E = torch.tensor([[1, 0],
                  [1, -1],
                  [2, 1]])

F = D.matmul(E)

print('D.matmul(E) =', F)
```

```python
F = D.mm(E)

print('D.mm(E) =', F)
```

```python
D = torch.tensor([[1, 1, 3],
                  [4, 5, 6],
                  [7, 8, 9]])

E = torch.tensor([[1, 0],
                  [1, -1],
                  [2, 1]])

F = D @ E

print('D @ E =', F)
```

```python
# 흑백 이미지의 대칭 이동
# 흑백 이미지의 좌우로 대칭 이동 코드 표현

G = torch.tensor([[255, 114, 140],
                  [39, 255, 46],
                  [61, 29, 255]])
print('G = ', G)
```

```python
import matplotlib.pyplot as plt

plt.xticks([]), plt.yticks([])
plt.imshow(G, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
H = torch.tensor([[0, 0, 1],
                  [0, 1, 0],
                  [1, 0, 0]])
print('H = ', H)
```

```python
I = G @ H

print('I = ', I)
```

```python
plt.xticks([]), plt.yticks([])
plt.imshow(I, cmap='gray', vmin=0, vmax=255)
plt.show()
```
