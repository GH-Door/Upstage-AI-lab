```python
from selenium import webdriver
browser = webdriver.Chrome() # 브라우저 다시 실행
```

```python
url = 'https://new.land.naver.com/complexes/111151?ms=35.161067,129.168383,17&a=APT:ABYG:JGC:PRE&e=RETAIL&ad=true'

browser.get(url)
```

```python
# 스크롤코드
import time
from selenium.webdriver.common.by import By

browser.find_element(By.CLASS_NAME, 'item_list').click()

# javascript코드를 통해서 스크롤을 내려준다. (네이버 로그인)
modal_elem = browser.find_element(By.CLASS_NAME, 'item_list')

import time
for i in range(10):
    js_code = 'arguments[0].scrollTo(0, arguments[0].scrollHeight)'
    browser.execute_script(js_code, modal_elem)
    time.sleep(2)
```

```python
## 전체 매물 데이터의 정보 (멀티 데이터) - 2시

# browser.find_element(By.CLASS_NAME, 'item')
datas = browser.find_elements(By.CLASS_NAME, 'item')
data_list = []

for i in datas:
    try:
        동정보 = i.find_element(By.CLASS_NAME, 'item_title').text
        가격 = i.find_element(By.CLASS_NAME, 'price_line').text
        매물정보 = i.find_element(By.CLASS_NAME, 'info_area').text
        
        data_list.append({
        '동' : 동정보,
        '가격' : 가격,
        '매물정보' : 매물정보
        })
    except:
        pass
```

```python
import pandas as pd

df = pd.DataFrame(data_list)
df.to_csv('네이버부동산.csv', encoding='utf-8-sig')
df
```

```python
import requests
import pandas as pd

def get_diningcode_restaurants(query="압구정동", order="r_score", page=1, size=20):
    """DiningCode API를 통해 특정 지역의 맛집 정보를 가져오는 함수"""
    
    url = "https://im.diningcode.com/API/isearch/"
    headers = {
        'Accept': 'application/json, text/plain, */*',
        'Accept-Language': 'ko,en;q=0.9,en-US;q=0.8',
        'Connection': 'keep-alive',
        'Content-Type': 'application/x-www-form-urlencoded',
        'Origin': 'https://www.diningcode.com',
        'Referer': 'https://www.diningcode.com/',
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)',
    }

    data = {
        'query': query,
        'order': order,
        'search_type': 'poi_search',
        'page': str(page),
        'size': str(size),
    }

    response = requests.post(url, headers=headers, data=data)
    
    if response.status_code == 200:
        return response.json()
    else:
        return None

def parse_restaurant_data(response_data):
    """API 응답 데이터를 가공하여 테이블 형태로 변환"""
    
    if not response_data or "result_data" not in response_data:
        print("데이터를 가져오지 못했습니다.")
        return None

    restaurants = response_data["result_data"]["poi_section"]["list"]
    
    # DataFrame 생성
    data = []
    for r in restaurants:
        data.append([
            r.get("nm", "N/A"),   # 식당 이름
            r.get("road_addr", "N/A"),  # 주소
            r.get("phone", "N/A"),  # 전화번호
            r.get("category", "N/A"),  # 음식 종류
            r.get("user_score", "N/A"),  # 평점
            r.get("review_cnt", 0),  # 리뷰 개수
            r.get("favorites_cnt", 0),  # 즐겨찾기 개수
            r.get("image", "N/A"),  # 이미지 URL
        ])

    df = pd.DataFrame(data, columns=["이름", "주소", "전화번호", "카테고리", "평점", "리뷰 수", "즐겨찾기 수", "이미지 URL"])
    return df

# 여러 페이지(1~5) 데이터 수집
query = "해운대"
order = "r_score"  # 평점순 정렬
size = 20  # 한 페이지당 20개 수집
all_data = []

for page in range(1, 6):  # 페이지 1~5
    response_data = get_diningcode_restaurants(query, order, page, size)
    df = parse_restaurant_data(response_data)
    if df is not None:
        all_data.append(df)

# 최종 데이터프레임 합치기
if all_data:
    final_df = pd.concat(all_data, ignore_index=True)

    final_df
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Selenium 웹드라이버 설정 (일반 모드)
driver = webdriver.Chrome()

# URL 접속
url = 'https://kream.co.kr/search?keyword=%EB%82%98%EC%9D%B4%ED%82%A4&tab=products&footer=home'
driver.get(url)

time.sleep(3)  # 페이지 로딩 대기

# 스크롤 내리기 (필요시 더 많은 데이터 로딩)
last_height = driver.execute_script("return document.body.scrollHeight")
for _ in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# 데이터 수집
items = driver.find_elements(By.CLASS_NAME, 'search_result_item')
print(f"총 {len(items)}개의 상품을 찾았습니다.")

data_list = []

for idx, item in enumerate(items):
    try:
        html = item.get_attribute('innerHTML')
        print(f"{idx + 1}번째 아이템 크롤링 중...")

        # 각 요소 안에서 필요한 데이터 추출
        # 브랜드
        brand = item.find_element(By.CLASS_NAME, 'brand-name').text.strip()
        # 제품명
        name = item.find_element(By.CLASS_NAME, 'name').text.strip()
        # 번역된 이름
        translated_name = item.find_element(By.CLASS_NAME, 'translated_name').text.strip()
        # 가격
        price = item.find_element(By.CLASS_NAME, 'amount').text.strip()
        # 거래 수
        try:
            trade_info = item.find_element(By.CLASS_NAME, 'status_value').text.strip()
        except:
            trade_info = ''
        # 관심 수 & 스타일 수
        try:
            action_info = item.find_element(By.CLASS_NAME, 'action_wish_review').text.strip()
        except:
            action_info = ''
        # 이미지 URL
        img_url = item.find_element(By.TAG_NAME, 'img').get_attribute('src')
        # 상품 상세 페이지 링크
        link = item.find_element(By.TAG_NAME, 'a').get_attribute('href')

        data_list.append({
            '브랜드': brand,
            '제품명': name,
            '번역명': translated_name,
            '가격': price,
            '거래수': trade_info,
            '관심/스타일': action_info,
            '이미지 URL': img_url,
            '링크': 'https://kream.co.kr' + link
        })
    except Exception as e:
        print(f"{idx + 1}번째 아이템에서 오류 발생: {e}")

# DataFrame으로 변환
df = pd.DataFrame(data_list)

# CSV 저장
df.to_csv('kream_nike_products.csv', index=False, encoding='utf-8-sig')
print("CSV 저장 완료! 'kream_nike_products.csv' 파일이 생성되었습니다.")

# 드라이버 종료
# driver.quit()
```

## 지그재그 수집

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Selenium 드라이버 실행 (일반 모드)
driver = webdriver.Chrome()

# URL 접속
url = 'https://zigzag.kr/search?keyword=%ED%95%98%EA%B0%9D%EB%A3%A9'
driver.get(url)
time.sleep(3)

# 스크롤 다운 (더 많은 상품 로드)
last_height = driver.execute_script("return document.body.scrollHeight")
for _ in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# 컨테이너 찾기
items = driver.find_elements(By.CLASS_NAME, 'product-card-root')
print(f"총 {len(items)}개의 상품을 찾았습니다.")

data_list = []

for idx, item in enumerate(items):
    try:
        print(f"{idx + 1}번째 상품 크롤링 중...")

        # 상품 상세 링크
        link_tag = item.find_element(By.CLASS_NAME, 'product-card-link')
        product_link = 'https://zigzag.kr' + link_tag.get_attribute('href')

        # 이미지 URL
        img_tag = item.find_element(By.TAG_NAME, 'img')
        img_url = img_tag.get_attribute('src')

        # 브랜드명
        brand = item.find_element(By.CLASS_NAME, 'zds4_1kdomr8').text.strip()

        # 상품명
        product_name = item.find_element(By.CLASS_NAME, 'zds4_1kdomrc').text.strip()

        # 할인율
        try:
            discount = item.find_element(By.CLASS_NAME, 'zds4_1jsf80i2').text.strip()
        except:
            discount = '없음'

        # 가격
        try:
            price = item.find_element(By.CLASS_NAME, 'zds4_1jsf80i3').text.strip()
        except:
            price = '가격 정보 없음'

        # 평점 및 리뷰
        try:
            rating = item.find_element(By.CLASS_NAME, 'zds4_1rbzv5c4').text.strip()
            review_count = item.find_element(By.CLASS_NAME, 'zds4_1rbzv5c5').text.strip()
        except:
            rating = '없음'
            review_count = '없음'

        data_list.append({
            '브랜드': brand,
            '상품명': product_name,
            '할인율': discount,
            '가격': price,
            '평점': rating,
            '리뷰수': review_count,
            '이미지 URL': img_url,
            '상세 링크': product_link
        })

    except Exception as e:
        print(f"{idx + 1}번째 상품에서 오류 발생: {e}")

# DataFrame으로 변환
df = pd.DataFrame(data_list)

# CSV 저장
df.to_csv('zigzag_hagyeongryuk_products.csv', index=False, encoding='utf-8-sig')
print("CSV 저장 완료! 'zigzag_hagyeongryuk_products.csv' 파일이 생성되었습니다.")

# 드라이버 종료
driver.quit()
```
